# cropping 

```python
# Load the image
data = np.array(plt.imread('./image/linkedin.png'))


# Set the red channel in this part of the image to 0
data[200:400, 200:400, 0] = 0

# Set the green channel in this part of the image to 255
data[200:400, 200:400, 1] = 0

# Set the blue channel in this part of the image to 0
data[200:400, 200:400, 2] = 0

# Visualize the result
plt.imshow(data)
plt.show()
```

# convert RGB to gray scale 

read more [here](https://stackoverflow.com/questions/12201577/how-can-i-convert-an-rgb-image-into-grayscale-in-python)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

img = mpimg.imread('./image/linkedin.png')     
gray = rgb2gray(img)    
plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```
